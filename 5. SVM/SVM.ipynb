```python
import numpy as np
from sklearn.datasets import make_blobs
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(101)
```

```python
num_features = 2
num_labels = 2
num_points = 150
```

```python
x, y = make_blobs(n_samples=num_points, n_features=num_features, centers=num_labels)
```

```python
plt.scatter(x[:,0], x[:,1], c = y)
```

```python
# def loss(labels, predictions, weights, lam):
#     return np.sum(1- labels * predictions) + (lam/2) * np.sum(weights**2)
```

```python
W1 = np.random.random()
W2 = np.random.random()
b = np.random.random()
#weights = np.array((W1, W2))
```

```python
lam = 0.001
learning_rate = 0.001
epochs = 1000

for epoch in range(epochs):
    for i in range(len(x)):
        y_hat = W1 * x[i,0] + W2 * x[i,1] + b
        error = (1 - y[i] * y_hat) + (lam/2) * (W1**2 + W2**2)
        if (y[i] * y_hat) < 1:
            W1 = W1 + learning_rate * (y[i]*x[i,0]) - (2*lam * W1)
            W2 = W2 + learning_rate * (y[i]*x[i,1]) - (2*lam * W2)
        else:
            W1 = W1 + learning_rate * (-2*lam * W1)
            W2 = W2 + learning_rate * (-2*lam * W2)
    print('Epoch: {}, Error: {}'.format(epoch, error))
```

```python
y_hat = W1 * x[:,0] + W2 * x[:,1]
y_hat_rounded = [1 if el >= 0.5 else 0 for el in y_hat]
```

```python
np.mean(y == y_hat_rounded)
```

```python
colors = ['b', 'g', 'r', 'c', 'm', 'y' ,'k']
```

```python
def color_points(labels):
    points_colors = [colors[int(label)] for label in labels]
    return points_colors
```

```python
points_colors = color_points(y)
```

```python
preds = np.array([[-4,0],[-5,-4]])
```

```python
plt.scatter(x[:,0], x[:,1], c = points_colors)
plt.scatter(preds[:,0], preds[:,1], marker = '^')
```

```python
y_preds = W1 * preds[:,0] + W2 * preds[:,1]
```

```python
y_preds_rounded = [1 if el >= 0.5 else 0 for el in y_preds]
```

```python
y_preds_colors = [colors[int(pred)] for pred in y_preds_rounded]
```

```python
plt.scatter(x[:,0], x[:,1], c = points_colors)
plt.scatter(preds[:,0], preds[:,1], marker = '^', c = y_preds_colors)
```

```python
def hyperplane(x,w1,w2,b,v):
    return (-w1*x-b+v) / w2
```

```python
datarange = (x.min()*0.9, x.max()*1.1)

hyp_x_min = datarange[0]
hyp_x_max = datarange[1]


db1 = hyperplane(hyp_x_min, W1, W2, -0.5, 0)
db2 = hyperplane(hyp_x_max, W1, W2, -0.5, 0)

plt.scatter(x[:,0], x[:,1], c = points_colors)
plt.scatter(preds[:,0], preds[:,1], marker = '^', c = y_preds_colors)
plt.plot([hyp_x_min, hyp_x_max], [db1, db2])
```
